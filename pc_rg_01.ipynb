### **Análisis de datos**

```
import pandas as pd
```

### **PyCaret**

```
from pycaret.regression import *
```

## **Conjunto de Datos**

```
url = "https://raw.githubusercontent.com/cristiandarioortegayubro/BA/main/Datasets/publicidad_multiple.csv"
```

## ***Dataframe***

```
datos = pd.read_csv(url, index_col = 0)
```

```
datos.head(10)
```

## **Configuración del entorno PyCaret**

```
regresion = setup(data = datos, #nombre del df
                  target = "ventas", #variable objetivo
                  train_size = 0.80, #tamaño de conjunto entrenamiento
                  session_id=123) #numero de valor semilla
```

```
regresion.dataset
```

```
regresion.dataset_transformed
```

## **Modelos**

```
models()
```

```
print(f"Los modelos de regresión son {len(models())}")
```

### **Comparando modelos**

```
#compare_models toma automaticamente la variable de tipo regresion
# en este caso la variable 'regresion' x eso no hace falta declararlo
modelos = compare_models() #compara los distintos modelos aplicables y cual es el mas eficiente en este caso
```
