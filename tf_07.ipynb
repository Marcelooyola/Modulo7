# **Carga de módulos y datos necesarios**

```
import pandas as pd
import plotly.express as px
import numpy as np
from scipy.stats import norm
```

# **Creación del DataFrame**

```
url="https://raw.githubusercontent.com/cristiandarioortegayubro/BA/main/Datasets/"
archivo="propiedades.csv"
regresion = pd.read_csv(url+archivo, index_col=0)
```

#**Preprocesamiento de datos**

```
regresion = regresion.drop(columns = ["id", "l1", "l2"])
regresion.sample(10)
```

```
regresion.info()
```

```
from pandas.core.frame import DataFrame

#numericas
cols_numericas = DataFrame(regresion.select_dtypes(exclude=["object"]))
cols_numericas
```

```
for i in cols_numericas.columns:
  media = cols_numericas[i].mean()
  sigma = 1
  x = np.linspace(media - 3*sigma, media + 3*sigma, 1000)
  y = norm.pdf(x, media, sigma)

  fig = px.line(x=x, y=y, labels={'x': 'Valor', 'y': cols_numericas[i].name}, title='Distribución Normal estándar ('+cols_numericas[i].name +')')

  # Mostrar el gráfico
  fig.show()
```
